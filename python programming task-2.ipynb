```python
class Solution(object):
    def defangIPaddr(self, address):
        """
        :type address: str
        :rtype: str
        """
        return address.replace(".", "[.]")
```

```python
class Solution(object):
    def findNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count = 0
        for num in nums:
            if len(str(num)) % 2 == 0:
                count += 1
        return count
```

```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count_map = {}
        good = 0        
        for num in nums:
            if num in count_map:
                good += count_map[num]
                count_map[num] += 1
            else:
                count_map[num] = 1
                
        return good
```

```python
class Solution(object):
    def smallerNumbersThanCurrent(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        sorted_nums = sorted(nums)
        lookup = {}
        for i, num in enumerate(sorted_nums):
            if num not in lookup:
                lookup[num] = i
        return [lookup[num] for num in nums]
```

```python
class Solution(object):
    def subtractProductAndSum(self, n):
        """
        :type n: int
        :rtype: int
        """
        s = str(n)
        product = 1
        sum = 0        
        for digit in s:
            d = int(digit)
            product *= d
            sum += d            
        return product - sum
        
```

```python
class Solution(object):
    def xorOperation(self, n, start):
        """
        :type n: int
        :type start: int
        :rtype: int
        """
        result = 0
        for i in range(n):
            current_num = start + 2 * i
            result ^= current_num      
        return result
        
```
